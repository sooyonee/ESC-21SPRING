```python
import scipy
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom
from scipy.stats import gamma

ya=[12,9,12,14,13,13,15,8,15,6]; #10=counts of type A
na=10;
yb=[11,11,10,9,9,8,7,10,6,8,8,9,7]; #13=counts of type B
nb=13;
sya=sum(ya); syb=sum(yb);

# prior parameters for theta ~ gamma(m,n) for theta_a, theta_b
am = 120 ;an = 10; bm=12; bn=1;

prior_a = st.gamma(am, scale= 1/an);
prior_b = st.gamma(bm, scale= 1/bn);
post_a= st.gamma(am+sya, scale= 1/(an+na));
post_b= st.gamma(bm+syb, scale= 1/(bn+nb));

```

**Group A Bayesian Analysis**

```python
#posterior mean, variance of A
print((am+sya)/(an+na))
print(post_a.mean())

print((am+sya)/((an+na))**2)
print(post_a.var())
```

```python
# posterior mean, variance of B
print((bm+syb)/(bn+nb))
print(post_b.mean())

print((bm+syb)/((bn+nb))**2)
print(post_b.var())
```

```python
# posterior 95% CI (Group A)
# upper and lower bounds
lbA=gamma.ppf(0.025, am+sya, scale=1/(an+na))
ubA=gamma.ppf(0.975, am+sya, scale=1/(an+na))
(lbA, ubA)
```

```python
# posterior 95% CI (Group B)
# upper and lower bounds
lbB=gamma.ppf(0.025, bm+syb, scale=1/(bn+nb))
ubB=gamma.ppf(0.975, bm+syb, scale=1/(bn+nb))
(lbA, ubA)
```

```python
thetas = np.linspace(0, 20, 300)
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(thetas, prior_a.pdf(thetas), label='A Prior', c='blue')
plt.plot(thetas,post_a.pdf(thetas), label='A Posterior', c='green')
plt.legend();
```
